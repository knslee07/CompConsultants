```python
df2=df.groupby('year')[main_cc_list].sum()
df2
```

# create a panel table for main CC

```python
df3=df2.stack().unstack('year')
df3.index.name='CC'
df3['total']=df3.sum(axis=1)
df3.sort_values(by='total', ascending=False)
```
