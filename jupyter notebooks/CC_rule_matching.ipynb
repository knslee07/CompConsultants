```python
def nlp_text(text):
    doc=nlp(text)
    for token in doc:
        print(token.text , token.pos_, token.tag_, token.dep_, token.ent_type_) # token has ent_type_ !!!!!
        if token.text=='consultant':
            print('Children of consultants are', list(token.children))
        if token.dep_=="ROOT":
            print('Children of ROOT are', list(token.children))
        #if token.text=="Strategies":
            #print('the head of Strategies is', token.head.text)
    return doc
        
```

```python
def find_main_CC(doc): 
    '''this function accepts a doc object and 
    searches 1. subject of the sentence, 2. CC, and 3. verb to figure out who hired which CC.'''
    s=set()
    t=set()

    regex1=re.compile(r'(committee)|(we)|(company)|(firm)', re.I)
    regex2=re.compile(r'(management)|\s(HR)\s?(department)?|(department)|(human resource)', re.I)
    regex3=re.compile(r"^[,.'&]$|^\s$|^(\s?co\.?)$|^(\s?inc\.?)$", re.I)
    #regex4=re.compile(r"(hire)|(retain)|(engage)")
    verbs=['retain', 'engage', 'hire', 'select', 'choose', 'retained', 'engaged', 'hired' ,'selected', 'chose', 'chosen']
    
    for i, sent in enumerate(doc.sents): # iterate by sentence
        print('\n\n' ,sent)
        if any([token.dep_=='neg' for token in sent]): # if a sentence has a negative
            continue
        if any([token.dep_=='nsubj' and regex1.search(token.text) for token in sent]): # If a subject is a committee
            #print([token.dep_=='nsubj' and regex1.search(token.text) for token in sent])
            print('committee is the subject in', i, 'sentence')
            for token in sent: 
                if token.ent_type_=="CC": #If there is a CC in the sentence # and (token.head.dep_=="ROOT" or token.head.head.dep_=="ROOT")
                    if any(token.text in verbs for token in sent if token.dep_=="ROOT"): # and the verb is one of the hiring ones.
                        print('the committee/company hired CC.')
                        for ent in sent.ents:
                            if token.text in ent.text and not regex3.search(token.text):# each token can be punctuation or whitespaces, too
                                print('CC token text is', token.text)
                                print(ent.text, 'added to s')
                                s.add(ent.text)
                                print(s)
                                continue

        # if any of the tokens in a sentence meets the condition
        elif any([token.dep_=='nsubj' and regex2.search(token.text) for token in sent]): # if a subject is a management
            print('management is the subject in', i, 'sentence')
            for token in sent: 
                if token.ent_type_=="CC": #If there is a CC in the sentence
                    #print('CC:', token.text, 'verb:', token.head.text)
                    if any(token.text in verbs for token in sent if token.dep_=="ROOT"):
                        print('management hires')
                    #if 'retain' in token.head.text or 'engage' in token.head.text or 'hire' in token.head.text:
                        for ent in sent.ents:
                            if token.text in ent.text and not regex3.search(token.text):
                                print('token text is', token.text)
                                print(ent.text, 'added to t')
                                t.add(ent.text)
                                print(t)
                                continue

    #passive voice

        elif any([token.dep_=='nsubjpass' and token.ent_type_=='CC' for token in sent]) or any(token.ent_type_=='CC' and token.head.dep_=='nsubjpass' for token in sent): 
            print('CC is the subject in', i, 'sentence')
            if any(token.text in verbs for token in sent if token.dep_=="ROOT"):
                print('CC is hired')
                if any(token.dep_=='pobj' and regex1.search(token.text) for token in sent): 
                    print('by the committee/company')
                    for token in sent:
                        for ent in sent.ents:
                            if token.text in ent.text and not regex3.search(token.text):# each token can be punctuation or whitespaces, too
                                print('token text is', token.text)
                                print(ent.text, 'added to s')
                                s.add(ent.text)
                                print(s)
                elif not any(token.dep_=='pobj' for token in sent):
                    print('there is no passive object.')
                    for token in sent:
                        for ent in sent.ents:
                            if token.text in ent.text and not regex3.search(token.text):# each token can be punctuation or whitespaces, too
                                print('token text is', token.text)
                                print(ent.text, 'added to s')
                                s.add(ent.text)
                                print(s)

                elif any(token.dep_=='pobj' and regex2.search(token.text) for token in sent): # by the management
                    print('by management')
                    print('ents of the sentence is', list(sent.ents))
                    for token in sent:
                        for ent in sent.ents:    
                            if token.text in ent.text and not regex3.search(token.text):# each token can be punctuation or whitespaces, too
                                print('token text is', token.text)
                                print(ent.text, 'added to t')
                                t.add(ent.text)
                                print(t)
    print(s, t)
    return s, t
```

```python
def find_data_CC(doc):
    '''finds data CC by matching patterns, returns a set of CC '''
    #call the matcher object 
    matches = matcher(doc)
    #the document object and it will return #match_id, start and stop indexes of the matched words
    s=set()
    for match_id, start, end in matches:
        string_id = nlp.vocab.strings[match_id]  
        span = doc[start:end]  # The matched span
        CC_data=doc[start]
        #print(CC_data.text)
        for ent in doc.ents:
            if CC_data.text in ent.text:
                #print(ent.text)
                s.add(ent.text)
    #else:
    #    print('CC_data is', CC_data)
    #print(match_id, string_id, start, end, span.text)
    #print(CC_data)
    #print(s)
    return s
```
